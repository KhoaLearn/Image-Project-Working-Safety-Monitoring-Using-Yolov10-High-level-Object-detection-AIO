```python
from pytube import YouTube

# Specify the YouTube video URL
YOUTUBE_VIDEO_URL = "https://youtu.be/wqPSsu7XQ74"

# Download the video
yt = YouTube(YOUTUBE_VIDEO_URL)
stream = yt.streams.get_highest_resolution()
video_path = stream.download(output_path='/content/', filename='downloaded_video.mp4')

print(f"Video downloaded to {video_path}")
```

```python
# Load the YOLOv10 model
MODEL_PATH = "yolov10n.pt"
model = YOLOv10(MODEL_PATH)

# Run the model on the downloaded video
cap = cv2.VideoCapture(video_path)

# Define the codec and create VideoWriter object
fourcc = cv2.VideoWriter_fourcc(*'XVID')
output_path = "/content/result_video.avi"
out = cv2.VideoWriter(output_path, fourcc, 20.0, (int(cap.get(3)), int(cap.get(4))))

# Process the video and save frames
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    results = model(frame)
    for result in results:
        boxes = result.boxes.data.cpu().numpy()
        for box in boxes:
            x1, y1, x2, y2 = map(int, box[:4])
            cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
    out.write(frame)

# Release everything
cap.release()
out.release()

print(f"Result video saved to {output_path}")
```
